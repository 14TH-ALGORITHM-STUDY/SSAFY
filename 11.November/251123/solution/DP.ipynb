#### [9095] 1, 2, 3 더하기

[9095 참고 블로그](https://psh0706.tistory.com/entry/%EB%B0%B1%EC%A4%80-15992-1-2-3-%EB%8D%94%ED%95%98%EA%B8%B0-1-python?category=1023245)

- n을 1, 2, 3의 덧셈 조합으로만 나타내 그 개수를 찾아야 함   
![9095-blog-1](./asset/9095-blog-1.png)

- n=3 일 때   
    - n=2 일 때 조합에 1을 추가   
    ![9095-blog-2](./asset/9095-blog-2.png)   

    - n=1 일 때 조합에 2를 추가   
    ![9095-blog-3](./asset/9095-blog-3.png)   

    - n=0 일 때 조합에 3을 추가   
    ![9095-blog-4](./asset/9095-blog-4.png)   

- 1, 2, 3만으로 n의 덧셈조합의 개수를 구하는 법
    - `n-1을 만드는 경우의 수`
    - `n-2를 만드는 경우의 수`
    - `n-3을 만드는 경우의 수`
    - 의 합

- n=4  
    - n=3 일 때 조합에 1 추가    
    ![9095-blog-5](./asset/9095-blog-5.png)   
 
    - n=2 일 때 조합에 2 추가    
    ![9095-blog-6](./asset/9095-blog-6.png)   

    - n=1 일 때 조합에 3 추가   
    ![9095-blog-7](./asset/9095-blog-7.png)  

- 결국 피보나치 st  
    - 점화식 : `DP[n] = DP[n-1] + DP[n-2] + DP[n-3]`  
    ![9095-blog-8](./asset/9095-blog-8.png)  
    

```python
T = int(input())
answer = [int(input()) for _ in range(T)]

dp = [0 for _ in range(max(answer)+1)]
# 아래의 로직이 n>3 일때 유효하므로 1,2,3을 미리 넣어줍니다.
dp[1] = 1
dp[2] = 2
dp[3] = 4

for i in range(4, len(dp)):
    dp[i] = dp[i-1] + dp[i-2] + dp[i-3]

for a in answer:
    print(dp[a])
```

#### [12101] 1, 2, 3 더하기 2

- n=3 일 때   
    - n=2 일 때 조합에 1을 **앞에** 추가

    - n=0 일 때 조합에 3을 **앞에** 추가

- n=4  
    - n=3 일 때 조합에 1을 **앞에** 추가    
    ![12101-blog-5](./asset/12101-blog-5.png)   
 
    - n=2 일 때 조합에 2를 **앞에** 추가    
    ![12101-blog-6](./asset/12101-blog-6.png)   

    - n=1 일 때 조합에 3을 **앞에** 추가   
    ![12101-blog-7](./asset/12101-blog-7.png)  

- DP 테이블은 이전과 동일  
    - 점화식 : `DP[n] = DP[n-1] + DP[n-2] + DP[n-3]`  
    ![12101-blog-8](./asset/12101-blog-8.png)  

- K 번째 조합 찾기   
    - K=3 일 때   
        ![12101-blog-9](./asset/12101-blog-9.png)   
        - 1~4번 째 : n-1 = 3 의 조합의 맨 앞에 1을 더해 만들어진 그룹 1
        - 5~6번 째 : n-2 = 2 의 조합의 맨 앞에 2를 더해 만들어진 그룹 2
        - 7번 째 : n-3 = 1 의 조합의 맨 앞에 3을 더해 만들어진 그룹 3
        - K=3 이므로 그룹 1에 속함 → K=3의 조합의 첫 번째 수는 1임을 알 수 있음
    - 2번 째 수 찾기 (1 + ? + ?)
        - 맨 앞의 1을 제외하면 n-1 = 3의 조합 중 K=3 번 째 조합을 찾는 문제가 됨  
        - 이제 n=3, K=3
        ![12101-blog-10](./asset/12101-blog-10.png)  
        (n=3인 조합의 수(DP[3])는 4)
        - 1~2번 째 : n-1 = 2 의 조합의 맨 앞에 1을 더해 만들어진 그룹 1
        - 3번 째 : n-2 = 1 의 조합의 맨 앞에 2를 더해 만들어진 그룹 2
        - 4번 째 : n-3 = 0 의 조합의 맨 앞에 3을 더해 만들어진 그룹 3
        - K=3 이므로 그룹 2에 속함 → K=3의 조합의 두 번째 수는 2임을 알 수 있음
    - 3번 째 수 찾기 (1 + 2 + ?)
        - 앞의 1과 2를 제외하면 n-2 = 1의 조합 중 K=3 번 째 조합을 찾는 문제가 됨?
        - K' = K - DP[n-1=2] = 3 - 2 = 1
        - 이제 n=1, K=1   
        ![12101-blog-11](./asset/12101-blog-11.png)   
        - 1번 째 : n-1 = 0의 조합의 맨 앞에 1을 더해 만들어진 그룹 1  
        - 단 한 개만 존재하기 때문에 K=3일 때 마지막 수는 1임을 알 수 있음   
    - 따라서 `1 + 2 + 1`이 답   

- K번 째 조합 찾기 정리
    ![12101-blog-12](./asset/12101-blog-12.png)  
    - n=4 일 경우
    - index=4, (n-1)=그룹 1, (n-2)=그룹 2, (n-3)=그룹 3
    - 이 때 K가 그룹 1, 2, 3 중에 어디에 속하는지 파악
        - K=3 → 그룹 1 → 첫 자리가 1 → result에 1 추가 → `result = 1`  
    - index=3 으로 옮겨 과정 반복  
    ![12101-blog-13](./asset/12101-blog-13.png)  
    - K=3 → 그룹 2 → 첫 자리가 2 → result에 2 추가 → `result = 1 + 2`  
    - index=1 (n-2)로 옮겨 과정 반복  
    ![12101-blog-14](./asset/12101-blog-14.png)  
    - K=3-2=1 → 첫 번째 조합은 n-1 그룹인 그룹 1에 포함되어야 하나 n=1 일 때는 0이 되므로 result에 1 추가 → `result = 1 + 2 + 1`      

```python
# 그룹 1에 속하는 경우

# k값이 1그룹(dp[index-1])개수 보다 작은 경우 
if index - 1 >= 0 and dp[index - 1] >= k:
    result += "1+"
    index -= 1
```

```python
# 그룹 2에 속하는 경우

# k값에서 1그룹 (dp[index-1])갯수를 뺀 값이 2그룹(dp[index-2])갯수보다 작으면 2그룹에 속한다.
if index - 2 >= 0 and dp[index - 2] >= (k - dp[index - 1]):
    result += "2+"
    k -= dp[index - 1]
    index -= 2
```

```python
# 그룹 3에 속하는 경우

# k값에서 1그룹 (dp[index-1])과 2그룹(dp[index-2]) 갯수를 뺀 값이
# 3그룹 (dp[index-3])보다 작으면 3그룹에 속한다.
if index - 3 >= 0 and dp[index - 3] >= (k - dp[index - 1] - dp[index - 2]):
    result += "3+"
    k -= (dp[index - 1] + dp[index - 2])
    index -= 3
```

```python
N, k = map(int, input().split())

dp = [1, 1, 2, 4]

if N > 3:
    i = 4
    while True:
        if i == N+1:
            break
        dp.append(dp[i-1] + dp[i-2] + dp[i-3])
        i += 1

if dp[N] < k:
    print(-1)
else:
    index = N
    result = ""
    while index > 0:
        if index - 1 >= 0 and dp[index - 1] >= k:
            result += "1+"
            index -= 1
        elif index - 2 >= 0 and dp[index - 2] >= (k - dp[index - 1]):
            result += "2+"
            k -= dp[index - 1]
            index -= 2
        elif index - 3 >= 0 and dp[index - 3] >= (k - dp[index - 1] - dp[index - 2]):
            result += "3+"
            k -= (dp[index - 1] + dp[index - 2])
            index -= 3

    print(result[:-1])
```
